```python
l = 1.0
b = 2
q_1 = 0.6
q_2 = 0.4
mu = 0.5 # 1/b

edge_equations = {
    '0.6λ': lambda p: p * q_1 * g.symbols['l'],
    '0.4λ': lambda p: p * q_2 * g.symbols['l'],
    'λ': lambda p: p * g.symbols['l'],
    '2μ': lambda p: p * 2 * g.symbols['mu'],
    'μ': lambda p: p * g.symbols['mu']
}

eqs = g.make_equations(edge_equations)
ps = g.solve(eqs, l=l, mu=mu)
```
